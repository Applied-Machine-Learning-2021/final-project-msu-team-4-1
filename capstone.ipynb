```python

import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, preprocessing
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
from nltk.metrics import edit_distance
from sklearn import datasets, preprocessing
from sklearn.svm import SVC

X = dfscaled.drop(['Spill'],axis=1).values
y = dfscaled['Spill'].values

# le = preprocessing.LabelEncoder()
# labels_encoded = le.fit_transform(df.iloc[:, 46])

scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# scaler.fit(dfscaled)
# print(scaler.transform(dfscaled))
# X_ = scaler.transform(dfscaled)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.30, random_state = 10)

# clf = MultinomialNB()
# clf.fit(abs(X), y)
# clf.fit(X_, y)
# predict = clf.predict(X_[0:1])
# print(accuracy_score(y,predict))
# X_[1, :]

```

```python
import math
from sklearn.metrics import mean_squared_error
#Create A MultinomialNB classifier
MN_mse=[]
MN_rmse=[]
MN_acc=[]

for alpha in range(1,5):
  clf = MultinomialNB(alpha=alpha, fit_prior=True)
  clf.fit(X_train, y_train)
  cl_pred = clf.predict(X_test)
  #Calculate accuracy of classifier based on test data
  accuracy = clf.score(X_test, y_test)
  MN_rmse.append(math.sqrt(mean_squared_error(y_test,cl_pred)))
  MN_mse.append(mean_squared_error(y_test,cl_pred))
  MN_acc.append(accuracy)
  
MN_mse_p = [i *100 for i in MN_mse]
MN_rmse_p = [i * 100 for i in MN_rmse]
MN_acc_p =  [i *100 for i in MN_acc]
per= range(0,4)
st = "Accuracy with (alpha="
#for i in per:
  #print( "Accuracy w.r.t to C")
  #print(st+ (str(i+1)) + "):" +(str(MN_acc_p[i])+ "%"))
  #print ("---------------------------------------------")
print( 'the best accuracy for MultinomialNB  was :' +(str(max(MN_acc_p))) + "%")
```

```python
#SVM classifier
sv_rmse =[]
sv_mse=[]
sv_acc =[]
for C in range(1,20):
  svm = SVC(kernel='linear',C=C, random_state=42)
  svm.fit(X_train, y_train)
  sv_pred = svm.predict(X_test)
  accuracy = svm.score(X_test, y_test)
  sv_rmse.append(math.sqrt(mean_squared_error(y_test,sv_pred)))
  sv_mse.append(mean_squared_error(y_test,sv_pred))
  sv_acc.append(accuracy)

sv_mse_p = [i *100 for i in sv_mse]
sv_rmse_p = [i * 100 for i in sv_rmse]
sv_acc_p =  [i *100 for i in sv_acc]
per= range(0,19)
st = "Accuracy with (C="
#for i in per:
  #print( "Accuracy w.r.t to C")
 # print (st+ (str(i+1)) + "):" +(str(sv_acc_p[i])+ "%"))
 # print ("---------------------------------------------")


print( 'The best accuracy for SVM was :' +(str(max(sv_acc_p))) + "%")
print( 'The lowest accuracy for SVM was :' +(str(min(sv_acc_p))) + "%")
print( 'The minimum error for SVM was :' +(str(min(sv_mse_p))) + "%")
print( 'The highest error for SVM was :' +(str(max(sv_mse_p))) + "%")
```

```python
#KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error
import math
K_rmse_all = []
K_mse_all = []
K_acc =[]
for n_neighbors in range(1,20):
  knr = KNeighborsClassifier(n_neighbors=n_neighbors, weights='uniform', algorithm='auto', leaf_size=30, p=2)
  knr.fit(X_train,y_train)
  knr_pred = knr.predict(X_test)
  knr_acc = knr.score(X_test,y_test)
  K_rmse_all.append(math.sqrt(mean_squared_error(y_test,knr_pred)))
  K_mse_all.append(mean_squared_error(y_test,knr_pred))
  K_acc.append(knr_acc)

knr_mse_p = [i *100 for i in K_mse_all]
knr_rmse_p = [i * 100 for i in K_rmse_all]
knr_acc_p =  [i *100 for i in K_acc]

print( 'The best accuracy for K Nearest Classifier was :' +(str(max(knr_acc_p))) + "%")
print( 'The lowest accuracy for K Nearest Classifier was :' +(str(min(knr_acc_p))) + "%")
print( 'The minimum error for K Nearest Classifier was :' +(str(min(knr_mse_p))) + "%")
print( 'The highest error for K Nearest Classifier was :' +(str(max(knr_mse_p))) + "%")
```
